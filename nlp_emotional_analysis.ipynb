```python
with open("story (2).txt", "r", encoding="utf-8") as f:
    text = f.read()

print(len(text))
```

```python
import textwrap

segments = textwrap.wrap(text, 500)

print("Number of segments:", len(segments))
print(segments[0])
```

```python
!pip install transformers
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
results = []

for i, segment in enumerate(segments[:200]):
    res = sentiment_pipeline(segment[:512])[0]

    results.append({
        "segment": i,
        "label": res['label'],
        "score": res['score']
    })

results[:5]
```

```python
import pandas as pd

df = pd.DataFrame(results)
df.head()
```

```python
df.to_csv("story_sentiment.csv", index=False)

from google.colab import files
files.download("story_sentiment.csv")
```
